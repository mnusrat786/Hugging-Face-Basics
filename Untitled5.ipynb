```python
import numpy as np
import matplotlib.pyplot as plt
from transformers import pipeline
from pylab import rcParams
```

```python
rcParams['figure.figsize'] = 10, 5
```

```python
zero_shot_classifier = pipeline("zero-shot-classification")
```

```python
result = zero_shot_classifier( sequences = "Can you order some Pizza & Coke to my software house?",
                              candidate_labels = ["Flight Travel", "Cabs Travel", "Reminders", "Food", "Movies"], multi_class= True )
```

```python
plt.bar(result["labels"], result["scores"])
plt.yticks(list(np.arange(0, 1, 0.1)))
plt.show()
```
